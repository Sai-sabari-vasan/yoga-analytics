```python
# Calculate the sum of all null values in the DataFrame
total_null_values = df.isnull().sum().sum()

# Print the total count of null values
print("Total count of null values in the DataFrame:", total_null_values)
```

columns with null values

```python
# Calculate the sum of missing values for each column
missing_values = df.isnull().sum()

# Filter out only columns with missing values
columns_with_missing_values = missing_values[missing_values > 0]

# Display columns with missing values and their corresponding counts
print(columns_with_missing_values)
```

getting all null value and their columns

```python
# Calculate the sum of null values for each column
missing_values = df.isnull().sum()

# Filter out columns with null values
columns_with_null_values = missing_values[missing_values > 0]

# Print the names of columns with null values
print("Columns with null values:")
for column_name, null_count in columns_with_null_values.items():
    print(f"Column '{column_name}' has {null_count} null value(s).")
```

visualization of null values

```python
# Calculate the sum of null values for each column
missing_values = df.isnull().sum()

# Filter out columns with null values
columns_with_null_values = missing_values[missing_values > 0]

# Plot the null values count in a bar chart
plt.figure(figsize=(10, 6))
columns_with_null_values.plot(kind='bar')
plt.title('Count of Null Values in Columns')
plt.xlabel('Columns')
plt.ylabel('Count of Null Values')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.tight_layout()
plt.show()
```

getting columns with less null values

```python
# Calculate the sum of null values for each column
missing_values = df.isnull().sum()

# Filter out columns with null values
columns_with_null_values = missing_values[(missing_values >0) & (missing_values<200)]

# Print the names of columns with null values
print("Columns with null values:")
for column_name, null_count in columns_with_null_values.items():
    print(f"Column '{column_name}' has {null_count} null value(s).")
```

```python
df1 = df.copy()  # Create a copy of the original DataFrame

# Columns to impute
columns_to_impute = [
    'Do you have any health problem?.1',
    'Give brief account of your physical and mental health status',
    'Family History-Mother',
    'Family History-Father',
    'Could you write briefly about what yoga means to you?',
    'What type of exercises are you involved in?',
    'How many months/years have you been practicing above mentioned exercises?',
    'How many hours per day do you watch movies or web series on T.V./laptop? ',
    'What type of recreation/hobbies do you have? Please specify:'
]

# Fill missing values in specified columns with the mean of each column
mean_values = df[columns_to_impute].mean()  # Calculate mean values for each column
df1[columns_to_impute] = df1[columns_to_impute].fillna(mean_values)
```
